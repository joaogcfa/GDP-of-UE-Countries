## Pergunta a ser respondida

Países que estão na UE possuem um crescimento em seu PIB

## Explicação da hipótese e mecanismo

Tendo em vista o fato de que a organização mundial "UE" é uma organização que busca estimular um progesso econômico dos paises que nesta estão, é esperado que ao entrar nessa organização o desenvolvimento econômico deles, com o tempo, seja aumentado.

## Bases de dados usadas

Os dados desse projeto foram retirados do site https://www.gapminder.org/data/, sendo que a partir desse pudemos ter os dados do crescimento do PIB por ano de todos os paises do mundo no seguinte site: https://www.rug.nl/ggdc/productivity/pwt/

## Análise de dados

**Importação de bibliotecas**

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

**Acolhimento dos dados do PIB**

```python
d1= pd.read_excel("gdp_total_yearly_growth.xlsx")
income=d1.set_index("country")
income.head()
```

**Criação de lista de países que estão na União Europeia**

```python
paises_ue=["France", "Germany", "Belgium", "Italy", "Luxembourg", "Netherlands", "Denmark", "Ireland", "United Kingdom", "Greece", "Portugal", "Spain", "Austria", "Finland", "Sweden", "Hungary", "Cyprus", "Czech Republic", "Estonia", "Latvia", "Lithuania", "Malta", "Poland", "Slovak Republic", "Slovenia", "Bulgaria", "Romania", "Croatia"]
```

**Criação de coluna de países que estão na União Europeia**

```python
esta_na_ue=[]
for pais in d1["country"]:
    if pais in paises_ue:
        esta_na_ue.append('sim')
    else:
        esta_na_ue.append('não')
income["Está na UE"]=esta_na_ue
income.head(10)
```

**Seleção de países que estão ou não na União Europeia**

```python
incomep=income.loc[["Germany","United Kingdom", "Greece", "France", "Canada", "Brazil", "China", "Mexico"]]
incomep
```

**Seleção de período relevante dado que a União Europeia foi criada em 1958**

```python
groth=incomep.iloc[:, 149:200].transpose()
groth.head()
```

**Calculo do crescimento do PIB mundial anual no mesmo período feito acima**

```python
world=income
world_gdp=world.iloc[:, 149:200].mean()
world_gdp.head()
```

# Plotagem de Gráficos

### Reino Unido

```python
plt.plot(groth["United Kingdom"], label="UK")
plt.plot(world_gdp, label="world")
plt.grid()
plt.legend()
plt.ylabel("Crescimento por ano em %")
plt.xlabel("Tempo em anos")
plt.title("Crescimento do Reino Unido e do mundo")
```

**Descrição do gráfico**: No gráfico vemos que até mais ou menos 1973 (ano que o Reino Unido entrou na organização) quase sempre a União Europeia crescia menos que o mundo até que a partir desse ano o país crescia mais que o mundo, mostrando, à priore, que ao entrar na União Europeia a economia do país melhorou

**Descrição do gráfico**: O caso da Grécia é praticamente o contrário do Reino Unido, já que vemos que até 1981(ano que o país entrou na União Europeia) o país parecia que tinha uma economia melhor, ao passo que a maioria do tempo esta crescia mais que o mundo, até que a Grécia entra na UE  e esse crescimento parece menor comparado à antes

# Linha temporal de diferença de mundo e UE

Para deixar mais visalmente agradável, replica-se a ideia dos gráficos anteriores, só que agora com apenas uma linha a qual representa a diferença do PIB de um país com o mundo, já que dessa forma é mais fácil perceber quando de fato os paises da União Europeia realmente estão evoluindo, porque mesmo que o PIB desses países crescer, não pode-se ignorar o fato de que o mundo cresce também e para responder a pergunta inicial temos que ver ser os países crescem além do que o mundo

```python
uk=groth["United Kingdom"]-world_gdp
plt.plot(uk)
plt.grid()
plt.ylabel("Crescimento por ano em %")
plt.xlabel("Tempo em anos")
plt.title("Difereça de Crescimento do Reino Unido e do mundo")
```

```python
greece=groth["Greece"]-world_gdp
plt.plot(greece)
plt.grid()
plt.ylabel("Crescimento por ano em %")
plt.xlabel("Tempo em anos")
plt.title("Difereça de Crescimento da Grécia e do mundo")
```

```python
france=groth["France"]-world_gdp
plt.plot(france)
plt.grid()
plt.ylabel("Crescimento por ano em %")
plt.xlabel("Tempo em anos")
plt.title("Difereça de Crescimento da França e do mundo")
```

```python
germany=groth["Germany"]-world_gdp
plt.plot(germany)
plt.grid()
plt.ylabel("Crescimento por ano em %")
plt.xlabel("Tempo em anos")
plt.title("Difereça de Crescimento da Alemanha e do mundo")
```

# BoxPlot

Para que se tivesse certeza de que entrar na União Europeia faz com que os paises que nela estão tem o seu PIB aumentado, foi feito 4 boxplots de 4 países da união Europeia com os dados da diferença entre o crescimento desses países e o mundo antes e depois de eles entrarem na organização. Dessa forma, foi possível perceber se de fato a economia dos paises melhoraram ao entrar na organização mundial.

### Reino Unido

```python
plt.boxplot([uk.iloc[0:24], uk.iloc[24:]])
```

**Descrição do boxplot:** Como esperado, podemos ver que ao entrar na União Europeia, o crescimento do Reino Unido aumentou, comparado ao mundo.

### Grécia

```python
plt.boxplot([greece.iloc[0:32], greece.iloc[32:]])
```

**Descrição do boxplot:** Também como previamente analizado, podemos ver que ao entrar na União Europeia, o crescimento da Grécia caiu, mas pode-se notar também que a mudança não é tão brusca.

### França

```python
plt.boxplot([france.iloc[0:9], france.iloc[9:]])
```

**Descrição do boxplot:** Vemos também que ao entrar na União Europeia, a França teve seu crescimento diminuido e parece estar mais vezes crescendo menos que o mundo, já que mais valores estão negativos no período que esta entrou na organização

### Alemanha

```python
plt.boxplot([germany.iloc[0:9], germany.iloc[9:]])
```

**Descrição do boxplot:** Vemos que a Alemanha é um claro exemplo de que entrar na União Europeia não faz com que sua economia avance, já que mostra que até entrar na organização, nunca o crescimento do paíse germânico era menor que o do mundo, já que não tem nenhum valor negativo e possui valores maiores que os valores da diferença de crescimento depois que este entra na União Europeia.
