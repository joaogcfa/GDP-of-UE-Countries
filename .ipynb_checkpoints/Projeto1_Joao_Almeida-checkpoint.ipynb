## Pergunta a ser respondida

Países que estão na UE tem o seu índice Gini dimunuido e seu PIB aumentado?

## Explicação da hipótese e mecanismo

Tendo em vista o fato de que a organização mundial "UE" é uma organização que busca estimular um progesso econômico dos paises que nesta estão, a desigualdade social dos Estados que participam na UE deveria dimuir com o tempo, já que agora os governos dos Estados que estão nessa organização possuem uma maior renda, ou seja, tem seu PIB aumentado, e dessa forma podem usar esse dinheiro para políticas públicas que melhorem o serviço básico das pessoas, dimunindo a desiguldade socail.


## Bases de dados usadas

Os dados desse projeto foram retirados do site https://www.gapminder.org/data/, sendo que a partir desse pudemos ter os dados do índice Gini( que é a forma de acompanhar a desigualdade social) e PIB de alguns países que estão na União Europeia

## Análise de dados

**Importação de bibliotecas**

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

**Acolhimento dos dados do índice Gini**

```python
dados = pd.read_excel('gini.xlsx')
gini=dados.set_index("country")
gini.head()
```



**Seleção dos países que estão na União Europeia**



```python
ginip = gini.loc[["United Kingdom","France","Germany","Greece"]]
ginip = ginip.transpose()
ginip.head()
```

**Seleção de uma faixa que corta os dados para um período de tempo relevante tendo em vista que a União Europeia é criada em 1958**

```python
faixa = (ginip.index >= 1950) & (ginip.index <= 2000)
ginip[faixa]
```

**Acolhimento dos dados do PIB**

```python
d1= pd.read_excel("income_per_person_gdppercapita_ppp_inflation_adjusted.xlsx")
income=d1.set_index("country")
income.head()
```

**Seleção dos países que estão na União Europeia**

```python
incomep = income.loc[["United Kingdom","France","Germany","Greece"]]
incomep = incomep.transpose()
incomep.head()
```

**Seleção de uma faixa que corta os dados para um período de tempo relevante tendo em vista que a União Europeia é criada em 1958**

```python
faixa2 = (incomep.index >= 1950) & (incomep.index <= 2000)
incomep[faixa2]
```

**Plotagem dos gráficos do Reino Unido**

Para a plotagem do gráfico do PIB e do índice Gini do Reino Unido, foi necessário dividir os valores do PIB por mil para que ambas as variaveis tivessem uma mesma escala e foi usada o PIB nesse caso so para comparar o comportamento desse ao índice Gini.

```python
plt.plot(ginip[faixa].loc[:,"United Kingdom"],label="gini")
plt.grid()
plt.plot((incomep[faixa2]/1000).loc[:,"United Kingdom"])
```

**Descrição do gráfico**: O Reino Unido entrou na União Europeia em 1973 e vemos que a partir desse momento o índice Gini aumenta, ao contrário do esperado, enquanto o PIB aumenta, conforme o esperado.

**Plotagem dos gráficos da Grécia**

Para a plotagem do gráfico do PIB e do índice Gini da Grécia, foi necessário dividir os valores do PIB por mil para que ambas as variaveis tivessem uma mesma escala e foi usada o PIB nesse caso só para comparar o comportamento desse ao índice Gini.

```python
plt.plot(ginip[faixa].loc[:,"Greece"])
plt.plot((incomep[faixa2]/1000).loc[:,"Greece"])
plt.grid()
```

**Descrição do gráfico**: A Grécia entrou na União Europeia em 1981 e vemos que a partir desse momento o índice Gini diminui, conforme o esperado, porém observamos que antes de eles entrarem na UE, a variável já estava diminuindo. Já o PIB aumenta, conforme o esperado.

**Plotagem dos gráficos da França**

Para a plotagem do gráfico do PIB e do índice Gini da França, foi necessário dividir os valores do PIB por mil para que ambas as variaveis tivessem uma mesma escala e foi usada o PIB nesse caso só para comparar o comportamento desse ao índice Gini.

```python
plt.plot(ginip[faixa].loc[:,"France"])
plt.plot((incomep[faixa2]/1000).loc[:,"France"])
plt.grid()
```

**Descrição do gráfico**: A França foi uma dos países que fundaram a UE, portanto esta está associada à organização desde 1958 e vemos que a partir desse momento o índice Gini aumenta um pouco e logo em seguida ele cai constamente, conforme o esperado. Além disso o PIB aumenta, conforme o esperado.

**Plotagem dos gráficos da Alemanha**

Para a plotagem do gráfico do PIB e do índice Gini da Alemanha, foi necessário dividir os valores do PIB por mil para que ambas as variaveis tivessem uma mesma escala e foi usada o PIB nesse caso só para comparar o comportamento desse ao índice Gini.

```python
plt.plot(ginip[faixa].loc[:,"Germany"])
plt.plot((incomep[faixa2]/1000).loc[:,"Germany"])
plt.grid()
```

**Descrição do gráfico**: A Alemanha foi uma dos países que fundaram a UE, portanto esta está associada à organização desde 1958 e vemos que a partir desse momento o índice Gini aumenta um pouco e logo em seguida ele cai constamente, conforme o esperado. Além disso o PIB aumenta, conforme o esperado.
